```python
from tripzilch.dsp import chunker, wav, wplot, midi
from scipy import ndimage, signal
from os import path

matplotlib.rcParams['figure.figsize'] = (18.0, 6.0)
wplot.rc_set_dark_colourscheme()

def widewindow(x, w):
    M = len(w) // 2
    r = x[:]
    r[:M] *= w[:M]
    r[-M:] *= w[-M:]
    return r

def sub_mean(x, N):
    N = int(N)
    L = len(x)
    y = zeros_like(x)
    ii = arange(-N, N + 1)
    k = 1.0 / len(ii)
    for n in range(L):
        iii = clip(ii + n, 0, L - 1)
        s = k * sum(x[iii])
        y[n] = x[n] - s
    print n, x[n], iii[0], iii[-1], s
    return y

def sub_poly(x):
    c = polyfit(arange(L), x.astype('float64'), 8)
    p = poly1d(c)(arange(L))
    plot(arange(L), p)
    return x - p
```
